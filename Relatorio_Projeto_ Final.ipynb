### *4.1 Existe algum padrão na distribuição de receita dos clubes ?*

```python
plt.figure(figsize=(10, 6))

data_to_plot = dadosf2.melt(id_vars=['Clube'],
             value_vars=['Valor_2023', 'Valor_2024'], var_name='Ano', value_name='Valor')

# Plotando o gráfico de barras
sns.barplot(data=data_to_plot, x='Clube', y='Valor', hue='Ano', palette='dark')

# Adicionando título e rótulos dos eixos
plt.title("Valorização dos times da Série A (2023-2024)", fontsize=20)
plt.xlabel("Clube", fontsize=14)
plt.ylabel("Valor", fontsize=14)

# Exibindo o gráfico
plt.legend(title='Ano', loc='upper right')
plt.xticks(rotation=45)  
plt.grid('darkgrid')
plt.tight_layout()
plt.show()
```

```python
dados_agrupados = dadosf1.groupby('tipo')[' receita'].sum().reset_index()

# Calculando o total de receitas
total_receitas = dados_agrupados[' receita'].sum()

# Valores para o tamanho de cada retângulo
sizes = dados_agrupados[' receita'].values
labels = dados_agrupados['tipo'].apply(lambda x: str(x)).values
percentages = [f"{(value/total_receitas)*100:.2f}%" for value in sizes]
new_labels = [f"{label}\n({percent})" for label, percent in zip(labels, percentages)]

# Cores - gerando uma paleta de cores
colors = plt.cm.Spectral(np.linspace(0, 1, len(labels)))

# Criando o TreeMap
plt.figure(figsize=(12, 8))
squarify.plot(sizes=sizes, label=new_labels, color=colors, alpha=0.8)

# Título do gráfico
plt.title('Distribuição de Receita por Tipo')
plt.axis('off')  # Desliga os eixos

# Mostrar o gráfico
plt.show()
```

No primeiro gráfico, é possível notar uma tendencia ascendente em relação à valorização dos clubes. Já no segundo gráfico, podemos perceber como ocorre a distribuição da receita dos mesmos. Pela ineficiência nas gestões esportivas, principalmente em distribuir de forma assertiva as suas receitas, alguns clubes acabam sendo dependentes marjoritários de direitos de tv e venda  de ativos (atletas) para manter o fluxo de caixa de seu clube.

### *4.3 Quantidade de torcedores influencia no desempenho esportivo?*

```python
import squarify

maiores_torcidas = dadost.groupby('Clube').sum()

cores_clubes = {
    'Flamengo': 'darkred',
    'Corinthians': 'grey',
    'São Paulo': 'pink',
    'Palmeiras': 'darkgreen',
    'Vasco': 'teal',
    'Cruzeiro': 'darkblue',
    'Grêmio': 'blue',  
    'Atlético-MG': 'yellow',
    'Bahia': 'lightblue',
    'Internacional': 'red',  
    'Fluminense': 'lightgreen',
    'Santos': 'darkgrey',
    'Botafogo': 'cyan',  
    'Sport': 'gold'
}

cores = [cores_clubes[clube] for clube in maiores_torcidas.index]

plt.figure(figsize=(10, 6))
squarify.plot(sizes=maiores_torcidas['Percentual (%)'], label=[f'{index}\n{value:.1f}%' for index, value in zip(maiores_torcidas.index, maiores_torcidas['Percentual (%)'])], color=cores, alpha=0.8)
plt.title('Maiores torcidas do Brasil', size=24)
plt.axis('off')

handles = [plt.Rectangle((0,0),1,1, color=cores_clubes[clube]) for clube in maiores_torcidas.index]
plt.legend(handles, maiores_torcidas.index, loc="upper left", fontsize='small', bbox_to_anchor=(1.05, 1))
plt.subplots_adjust(left=0.1, right=0.85)  
plt.show()
```

```python
def plot_valor_medio_e_variancia(dados):
    # Calcule o valor médio e a variância de cada time
    valor_medio_por_time = dados.groupby('Clubes')['Valor_total'].mean().sort_values(ascending=False)
    variancia_por_time = dados.groupby('Clubes')['Valor_total'].var().sort_values(ascending=False)
    
    # Crie o gráfico de barras para o valor médio
    fig, ax1 = plt.subplots(figsize=(12, 6))
    ax1.bar(valor_medio_por_time.index, valor_medio_por_time, color='blue')
    ax1.set_xlabel('Time')
    ax1.set_ylabel('Valor Médio', color='blue')
    ax1.tick_params('y', colors='blue')
    ax1.set_xticks(range(len(valor_medio_por_time.index)))  # Especificando os locais dos ticks
    ax1.set_xticklabels(valor_medio_por_time.index, rotation=45, ha='right')
    
    # Adicione o gráfico de linhas para a variância
    ax2 = ax1.twinx()
    ax2.plot(variancia_por_time.index, variancia_por_time, color='orange', marker='o', linestyle='-', linewidth=2)
    ax2.set_ylabel('Variância', color='orange')
    ax2.tick_params('y', colors='orange')
    
    plt.title('Valor Médio e Variância de cada time')
    plt.grid(True)
    plt.tight_layout()
    plt.show()


plot_valor_medio_e_variancia(dados) 
```

Geralmente há um mito no futebol em que " peso da camisa " ganha jogos, no entanto baseado na análise o principal fator para aumento do desempenho esportivo passa pela profissionalização das gestões esportivas, pois clubes tradicionais e com grande número de torcedores como por exemplo : Corinthians, Vasco e Cruzeiro, possuem um baixo desempenho financeiro e esportivo nos últimos 5 anos, mesmo possuindo uma grande torcida.
Em contrapartida, equipes como Fortaleza e Athletico Paranaense são exemplos de gestões profissionais que mesmo com poucos torcedores e poucas receitas comparadas aos clubes de maiores torcidas, eles( Fortaleza e Athletico) possuem equipes competitivas e hoje são referências em gestão e análise de desempenho na América Latina.

## 5. Conclusão

Ao examinar os padrões observados nos clubes e a situação financeira do futebol brasileiro, algumas constatações relevantes emergem.

Primeiramente, evidencia-se a presença de determinados padrões nos clubes, que podem influenciar tanto o desempenho esportivo quanto a saúde financeira. Porém, é notável que muitos clubes brasileiros não aderem a esses padrões, enfrentando dificuldades decorrentes de gestão ineficiente e dependência excessiva de fontes de receita específicas, como direitos de TV e venda de jogadores.

Além disso, é importante destacar que o "peso da camisa" não é mais garantia de sucesso no futebol atual. O desempenho esportivo está cada vez mais relacionado à profissionalização da gestão esportiva. Clubes tradicionais, como Corinthians, Vasco e Cruzeiro, apesar de sua grande base de torcedores, têm enfrentado dificuldades tanto financeiras quanto esportivas nos últimos anos. Por outro lado, equipes como Fortaleza e Athletico Paranaense têm se destacado pela excelência na gestão e pelo desempenho esportivo consistente, mesmo com recursos mais limitados e menor base de torcedores.

Assim, torna-se claro que a profissionalização da gestão é um elemento crucial para o sucesso sustentável no futebol brasileiro. A implementação de práticas de gestão eficientes e a diversificação das fontes de receita são fundamentais para que os clubes possam superar desafios financeiros e esportivos, tornando-se referências não apenas no cenário nacional, mas também internacionalmente.
