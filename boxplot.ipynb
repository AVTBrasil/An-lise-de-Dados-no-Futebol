```python
import pandas as pd

caminho = r"c:\Users\anton\Documents\tabela_clubes.csv"

df = pd.read_csv(caminho)


df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
colunas = ['Unnamed: 13','Unnamed: 14','Unnamed: 15','Unnamed: 16']
df2 =  df.drop(columns=colunas)
df2.head()
```

Tarefas
1. Criar um Jupyter Notebook para iniciar a exploração dos conjuntos de
dados
2. No notebook, calcular medidas de centralidade e dispersão das variáveis
disponíveis
3. No notebook, criar boxplots para as variáveis disponíveis


```python
# Calcular média e desvio padrão
medias = df2.mean()
desvios_padrao = df2.std()

# Exibir as medidas de centralidade
print("Médias:")
print(medias)
print("\nDesvios Padrão:")
print(desvios_padrao)
```

```python
# Criar histograma para 'Vitorias'
plt.hist(df2['Vitorias'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Vitórias')
plt.ylabel('Frequência')
plt.title('Histograma de Vitórias')
plt.show()
```

```python
# Criar boxplot para 'Saldo'
plt.boxplot(df2['Saldo'])
plt.xlabel('Saldo')
plt.title('Boxplot de Saldo')
plt.show()
```

```python
# Exemplo de criação de boxplot para a coluna 'Vitorias'
plt.figure(figsize=(8, 6))
plt.boxplot(df2['Vitorias'])
plt.title('Boxplot de Vitórias')
plt.ylabel('Número de Vitórias')
plt.show()
```

```python
vitorias_por_clube = df2.groupby('Clubes')['Vitorias'].sum()

# Ordenar os clubes pelo número de vitórias em ordem decrescente
clubes_ordenados = vitorias_por_clube.sort_values(ascending=False)

# Obter os três clubes com o maior número de vitórias
tres_melhores_times = clubes_ordenados.head(3)

print("Os três times com o maior número de vitórias:")
print(tres_melhores_times)
```

```python
tres_melhores_times = pd.DataFrame({
    'Clubes': ['Time1', 'Time2', 'Time3'],  # Substitua 'Time1', 'Time2', 'Time3' pelos nomes reais dos times
    'Vitorias': [100, 95, 90]  # Substitua 100, 95, 90 pelos números reais de vitórias dos times
})

# Criar um boxplot com os dados dos três times
plt.figure(figsize=(8, 6))
plt.boxplot(tres_melhores_times['Vitorias'])
plt.title('Boxplot do Número de Vitórias dos Três Melhores Times')
plt.ylabel('Número de Vitórias')
plt.xticks([1], ['Três Melhores Times'])
plt.grid(True)
plt.show()
```

```python
dados_time = df2[df2['Clubes'] == 'Gremio']

# Criar o boxplot das vitórias ao longo do tempo
plt.figure(figsize=(10, 6))
plt.boxplot(dados_time['Vitorias'], labels=['Gremio'])
plt.title(f'Boxplot das Vitórias de Gremio ao Longo do Tempo')
plt.xlabel('Time')
plt.ylabel('Vitórias')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.boxplot(df2['Vitorias'], vert=False)
plt.title(f'Boxplot de Vitórias de Gremio por Ano')
plt.xlabel('Vitórias')
plt.ylabel('Ano')
plt.yticks([1], ['Gremio'])  # Nome do time no eixo y
plt.grid(True)
plt.show()
```
